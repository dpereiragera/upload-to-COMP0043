```python
import numpy as np
import pandas as pd
from scipy.special import erf
from scipy.stats import norm
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import lognorm
from pytictoc import TicToc
from scipy.stats import ncx2
```

```python
#Plot the variance of a Cox-Ingersoll-Ross process
#dX = alpha*(mu-X)*dt + sigma*sqrt(X)*dW

#Settings
mu,sigma,alpha=0.07,0.265,5 #model parameters
X0=0.03 #initial value
t=np.arange(0,2.01,0.01) #time grid

#See e.g. PDE Tools page 64
var =X0*sigma**2/alpha*(np.exp(-alpha*t)-np.exp(-2*alpha*t)) + mu*sigma**2/(2*alpha)*(1-np.exp(-alpha*t))**2
var0 = X0*sigma**2*t
varinfty = mu*sigma**2/(2*alpha)*np.ones(np.size(t))

#Do the plot
plt.figure(1)
plt.plot(t,var,'r',t,var0,'g',t,varinfty,'b')
plt.legend(['var','var_O','var_infty'])
plt.xlabel('t')
plt.ylabel('Var(X)')
plt.ylim([0,6e-4])
plt.xlim([0,2])
plt.title('Variance of a Cox-Ingersoll-Ross process dX = \alpha(\mu-X)dt + \sigmaX^{1/2}dW')
plt.show()
```
