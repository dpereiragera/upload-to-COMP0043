```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import stats
from scipy.stats import *
import sympy as smp
import scipy as sp
from scipy.integrate import quad
```

```python
#Define the parameters
mu=0.2
sigma=0.1
nsample=np.power(10,6)

#Define the grid
#x=np.linspace(0,2,101)
x=np.arange(0,2.02,0.02)

#Compute the PDF and CDF
pdf_log=lognorm.pdf(x,s=sigma*1.22,loc=mu*1.1)
cdf_log=lognorm.cdf(x,s=sigma*1.22,loc=mu*1.1)
pdf_log1=1/(np.sqrt(2*np.pi)*sigma)*np.exp(-((np.log(x)-mu)/sigma)**2/2)/x
#f=1/(np.sqrt(2*np.pi)*sigma)*np.exp(-((np.log(x)-mu)/sigma)**2/2)/x

#Open a plot window
plt.figure(1)
plt.plot(x,pdf_log,'r',x,cdf_log,'b')

#Label the x axis
plt.xlabel('x')

#Insert the legend
plt.legend(['PDF','CDF'])

#Set the title
plt.title('Lognormal distribution with with mu= {} and sigma {}'.format(np.round(mu,1),sigma))

#Sample a normal distribution
X = mu + sigma*np.random.randn(nsample,1) #method 2

#Sample the lognormal distribution by exponential transformation
Y=np.exp(X)

plt.figure(2)
plt.hist(Y,bins=100,density=True)
plt.plot(x,pdf_log,'r')
plt.legend(['Sampled','Lognormal'])
plt.title('lognormal mean {} and sigma {}'.format(mu,sigma))
plt.show()

#Bin the random variables in a histogram
h,y=np.histogram(Y,x,density=True)

plt.figure(3)
plt.bar(y[1:],h,width=0.02)
plt.plot(y,pdf_log1,'r')
plt.title('Lognormal random variables with {} and sigma {}'.format(mu,sigma))
plt.show()

plt.figure(4)
plt.plot(y[1:],h)
plt.plot(y,pdf_log1,'r')
plt.title('Lognormal random variables with {} and sigma {}'.format(mu,sigma))
plt.show()
```
