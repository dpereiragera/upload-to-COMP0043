```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import stats
from scipy.stats import *
import sympy as smp
import scipy as sp
from scipy.integrate import quad
from scipy.special import erf
from scipy.special import iv
```

```python
#Fibonacci random number generator (Knuth)
#Seydel, Course Notes, Chapter 2, page 211-212

# Parameters
mu=5 
nu=17
M=714025
seed=12345
nsample=np.power(10,4)

# Linear congruential generator to start the sequence
a,b=1366,150889
U=np.zeros(nsample)
U[0]=seed

for i in range(1,nsample):
   U[i]=np.mod(a*U[i-1]+b,M)

#Fibonacci generator
Us=[]
for j in range(nu+1,nsample):
    calc=np.mod(U[j-nu]-U[j-mu],M)/M
    Us.append(calc)
U=Us

#Tests 

#Bin the random variables in a histogeam and normalise it 
nbins=100
h,x=np.histogram(U,nbins)

#Plot of the probability density function
plt.figure(1)
plt.hist(U,nbins,density=True)
plt.plot(x,np.ones(np.size(x)))
plt.legend(['Sampled','Theory'])
plt.ylim((0,2))
plt.xlim((-0.2,1.2))
plt.xlabel('x')
plt.ylabel('f')
plt.title('Uniform distribution')
plt.show()
```

```python
#Scatter plot
plt.figure(2)
plt.scatter(x=U[0:-2],y=U[1:-1])
plt.xlabel('U_{i-1}')
plt.xlabel('U_i')
plt.title('2D scatter plot')
plt.show()
```
