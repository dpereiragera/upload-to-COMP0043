```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import stats
from scipy.stats import *
import sympy as smp
import scipy as sp
from scipy.integrate import quad
from scipy.special import erf
from scipy.special import iv
```

```python
#Plot and sample the exponential distribution

#Define the parameters
mu=(2/3) #mean
nsample=np.power(10,6) #number of random samples

#Define the grid
#x=np.arange(0,5.05,0.05)
x=np.linspace(0,5,101)

#Compute the PDF and the CDF
#pdf_exp=(1/mu)*(np.exp(-x/mu)) 
#cdf_exp=1-np.exp(-x/mu) 
pdf_exp=expon.pdf(x,loc=0,scale=mu)
cdf_exp=expon.cdf(x,loc=0,scale=mu)

#Open a plot window
plt.figure(1)
plt.plot(x,pdf_exp)
plt.plot(x,cdf_exp)

#Lable the x axis
plt.xlabel('x')
plt.ylabel(['distributions'])

#Insert the legend
plt.legend(['PDF','CDF'])

#set the title
plt.title('Exponential distribution with mu= {}'.format(np.round(mu,1)))
plt.show()

#Sample an exponential distribution
U = np.random.rand(nsample,1) #standard uniform random numbers
X = -mu*np.log(U) #transformation to exponential random numbers

#Bin the random variables in a histogram and normalise it
dx=x[2]-x[1] #bin width
h,xx=np.histogram(X,x+dx/2)
h=h/(nsample*dx) #normalisation

plt.figure(2)
plt.plot(xx[1:],h,'b',x,pdf_exp,'r')
plt.xlim([0,5])
plt.xlabel('x')
plt.ylabel('pdf_exp')
plt.legend(['Sampled','Theory'])
plt.title('Exponential distribution with mu={}'.format(np.round(mu,1)))
plt.show()

plt.figure(3)
plt.bar(xx[1:],h,width=0.05)
plt.plot(x,pdf_exp,'r')
plt.xlabel('x')
plt.ylabel('pdf_exp')
plt.legend(['Sampled','Theory'])
plt.title('Exponential distribution with mu={}'.format(np.round(mu,1)))
plt.show()
```
