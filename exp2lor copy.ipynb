```python
import numpy as np
import pandas as pd
from scipy.special import erf
from scipy.stats import norm
import scipy as sp
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import lognorm
import threading
from threading import Thread
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
#Check numerically the Fourier pair Laplace <-> Lorentzian

N=1024 #grid size
Dx=0.1 #grid step in real space
x=Dx*np.arange(-N/2,(N/2),1) #grid in real space
Lx=N*Dx #upper truncation limit in real space
Dnu= (2*np.pi)/Lx #grid step in Fourier space 
nu=Dnu*np.arange(-N/2,(N/2),1)
Nu=N*Dnu # =2*np.pi/Dx

a=1
fa=(a/2)*np.exp(-a*np.abs(x)) #Laplace (or double exponential)
Fa=(a**2)/((a**2)+nu**2) #Lorentzian (or Cauchy)

Fn= np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(fa)))*Lx
fn= sp.fft.fftshift(np.fft.fft(np.fft.ifftshift(Fa)))/Lx
Fn1= np.fft.fftshift(np.fft.fft(np.fft.ifftshift(fa)))*Dx
fn1= np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(Fa)))/Dx
```

```python
plt.figure(1)
plt.plot(x,np.real(fn),'r')
plt.plot(x,abs(np.imag(fn)),'g')
plt.plot(x,fa,'k:')
plt.axis([-10,10,0,1])
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Re(fn','Im(fn)','fa'])
plt.show()
```

```python
plt.figure(2)
plt.plot(nu,np.real(Fn),'b')
plt.plot(nu,abs(np.imag(Fn)),'m')
plt.plot(nu,Fa,'c:')
plt.axis([-20,20,0,1])
plt.xlabel('omega')
plt.ylabel('F')
plt.legend(['Re(Fn)','Im(Fn)','Fa'])
plt.show()
```

```python
plt.figure(3)
plt.plot(x,np.real(fn1),'r')
plt.plot(x,abs(np.imag(fn1)),'g')
plt.plot(x,fa,'k:')
plt.axis([-10,10,0,1])
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Re(fn1)','Im(fn1)','fa'])
plt.show()
```
