```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import stats
from scipy.stats import *
import sympy as smp
import scipy as sp
from scipy.integrate import quad
from scipy.special import erf
from scipy.special import iv
```

```python
#Plot and sample the distribution of X = cos U

ngrid=200
nsample=np.power(10,4)

# (a) Compute the PDF
dx= 1/ngrid
x=np.linspace(0,1,ngrid)
f=2/(np.pi*np.sqrt((1-x**2)))

# (b) Sample
dy= np.pi/ngrid
y=np.linspace(-np.pi/2,np.pi/2,ngrid)
U= np.pi*(np.random.randn(nsample)-0.5)
hu=np.histogram(U,y,density=True)
hu=hu[0]
X= np.cos(U)
hx=np.histogram(X,x,density=True)
hx=hx[0]

#Plot the PDF of U
plt.figure(1)
plt.plot(y[1:],hu,'r',y,np.ones(np.size(y))/np.pi,'b')
plt.ylim([0,1])
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Sampled','Theory'])
plt.title('Distribution of U')
plt.show()

#Plot the PDF of X
plt.figure(2)
plt.plot(x[1:],hx,'r',x,f,'b')
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Sampled','Theory'])
plt.title('Distribution of X= cos U')
plt.show()
```
