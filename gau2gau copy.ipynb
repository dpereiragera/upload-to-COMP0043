```python
import numpy as np
import pandas as pd
from scipy.special import erf
from scipy.stats import norm
import scipy as sp
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import lognorm
import threading
from threading import Thread
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
#Check numerically that Gaussians form a closed set with respect to the Fourier transform

N=1024
Dx=0.1
x=Dx*np.arange(-N/2,N/2,1)
L=N*Dx
Dxi=2*np.pi/L #Nyquist relation: Dx*Dxi = 2*np.pi/N
xi=Dxi*np.arange(-N/2,N/2,1)
W=N*Dxi #=2*np.pi/Dx

a=1
fa=np.sqrt(a/np.pi)*np.exp(-a*x**2) #Gaussian in x space
Fa=np.exp(-xi**2/(4*a)) #Gaussian in xi space
Fn=np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(fa)))*L
fn=np.fft.fftshift(np.fft.fft(np.fft.ifftshift(Fa)))/L
Fn1=np.fft.fftshift(np.fft.fft(np.fft.ifftshift(fa)))*Dx
fn1=np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(Fa)))/Dx

plt.figure(1)
plt.plot(x,np.real(fn),'r')
plt.plot(x,abs(np.imag(fn)),'g')
plt.plot(x,fa,'y:')
plt.axis([-5,5,0,1])
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Re(fn)','Im(fn)','fa'])
plt.show()
```

```python
plt.figure(2)
plt.plot(xi,np.real(Fn),'b')
plt.plot(xi,abs(np.imag(Fn)),'m')
plt.plot(xi,Fa,'c:')
plt.axis([-10,10,0,1])
plt.xlabel('xi')
plt.ylabel('F')
plt.legend(['Re(Fn)','Im(Fn)','Fa'])
plt.show()
```

```python
plt.figure(3)
plt.plot(x,np.real(fn1),'r')
plt.plot(x,abs(np.imag(fn1)),'g')
plt.plot(x,fa,'y:')
plt.axis([-5,5,0,1])
plt.xlabel('x')
plt.ylabel('f')
plt.legend(['Re(fn1)','Im(fn1)','fa'])
plt.show()
```
