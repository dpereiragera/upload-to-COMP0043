```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import stats
from scipy.stats import *
import sympy as smp
import scipy as sp
from scipy.integrate import quad
from scipy.special import erf
from scipy.special import iv
```

```python
#Linear congruential random number generator
#Seydel, Course Notes, Chapter 2, page 202

#Good example, page 203
a,b,M=1597,51749,244944
seed=12345
nsample=np.power(10,4)
nbins=100

#Pathologic example, page 208
# a,b,M=2,0,11
# seed=1
# nsample=11

#Pathologic example, page 209
# a,b,M=1229,1,2048
# seed=1
# nsample=2048

#Generator
U=np.zeros(nsample)
U[0]=np.mod(seed,M) 
for i in range(1,nsample):
    U[i]=np.mod(a*U[i-1]+b,M)
U=U/M

#Quality checks

#Probability density function
plt.figure(1)
h,x=np.histogram(U,nbins,density=True)
plt.plot(x,np.ones(np.size(x)))
plt.hist(U,nbins,density=True)
plt.xlim([-0.2,1.2])
plt.ylim([0,2])
plt.xlabel('x')
plt.ylabel('f_U')
plt.legend(['Sample','Theory'])
plt.title('Probability density function')
plt.show()
```

```python
#2D scatter plot
plt.figure(2)
plt.scatter(x=U[0:nsample-1],y=U[1:nsample])
plt.xlabel('U_i')
plt.ylabel('U_{i+1}')
plt.title('2D scatter plot')
plt.show()
```

```python
#3d scatter plot
fig = plt.figure(3)
ax = fig.add_subplot(projection='3d')
ax.scatter(U[0:nsample-2],U[1:nsample-1],U[2:nsample])
ax.set_xlabel('U_i')
ax.set_ylabel('U_{i+1}')
ax.set_zlabel('U_{i+2}')
ax.set_title('3D scatter plot')
plt.show()
```
